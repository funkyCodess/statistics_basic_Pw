**1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval and ratio scales.**


---
**Qualitative Data** : It is a form of descriptive data which is generally non numerical in nature upon which arithematic operations such as addition, subtraction, multiplication or division.

eg:
* Ranks of students in a class.
* Addresses of teachers in a school.
* Gender of workers in a factory.


**Quantitative Data**: It is a form of data which represents a numerical value for the parameter of our data, that can be added, subtracted, multiplied or divided to.

eg:
* Salary of employees.
* Marks of students.
* Population of cities.


**SCALES OF MEASUREMENT**
1. **Nominal** : It is a scale of measurement which is used for data that is qualitative in nature and has no particular order inherent to it. It can be represented as pie charts, bar plots, etc. eg : Ranks of students in a class, Gender, Ethnicity.


2. **Ordinal** : It is a scale of measurement which is used for data that is qualitative in nature but has an inherent order to it. Difference cannot be measured. It can be represented as pie charts, bar plots. eg : Salaries of employees marked as high, medium, low.

3. **Interval** : It is a scale of measurement used for data in which order and rank is important. Difference can be measured between values with the exception of ratios. It doesn't have compulsory a 0 starting value due to lack of ratios. eg : Scores, Heights, Weights, Temperature.

4. **Ratio** : It is a scale of measurement used for data in which order and rank are of importance. In this scale, difference and ratios are measureable but it needs to have a 0 starting value. eg : Heights, Weights, Scores but not temperature(it can be negative and thus ratios are not meaningful).









**2. What are the measures of central tendencies and when should you use each? Discuss the mean, median, mode with examples and situations where each is appropriate.**


---

**Central tendencies** are the values that represent the data as a whole in order to gain some insights. There are 3 measures of central tendencies :     

* **Mean** : It is also known as the arithematic mid value of data. It can be calculated as the sum of all observations divided by the number of total observations. eg:
Let the marks of 5 students in a class (out of 100) be represented as [90, 78, 35, 67, 80]. Its mean can be calculated as -- Sum = 90+78+35+67+80 = 350.
Therefore MEAN = 350/5 = 70.
It can be used to calculate average scores, average incomes, etc. Mean is susceptible to outliers.


* **Median** : It is also known as the physical mid point of data. It is only calculated over sorted data.
For odd number of values, the median is the middle element.
For even number of values, the median is the average of the two middle elements.
It can be used instead of mean when there are outliers in our data.
It can be used for household incomes where few high earning families can act as outliers in order to skew the actual representaion of data.
Example:
    * For even dataset : 10 , 20, 30, 40
    
    Median = (average of middle terms) = (20 + 30)/2 = 25
    * For odd dataset : 10, 20, 30, 40, 50

    Median = (middle element) = 30


* **Mode** : It is the most occuring value in the dataset.
It can be used to find the most popular product.
It is independent to outliers.
May not exist when all the data values are unique in our dataset. eg :    
    * Mode(10, 20, 20, 20, 30, 20) = 20 (Most occuring)
    * Mode(10, 20, 10, 30, 20) = 10 and 20 (Most occuring) (bi-modal)


    
    





**3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?**


---
**Dispersion** refers to the degree to which data points in a dataset are spread out or scattered. It provides insights into the variability or spread of the data around a central value (mean, median, or mode). While measures of central tendency (mean, median, mode) summarize the center of the data, measures of dispersion explain how much the data varies.


**VARIANCE** is the average of the squared differences between each data point and the mean. It measures the overall variability of the data.

variance.png

Example : 2,4,6,8,10. Mean = 6. Variance : [(2-6)^2 + (4-6)^2 + (6-6)^2 + (8-6)^2 + (10-6)^2]/5 = 8.

**STANDARD DEVIATION** is the square root of the variance. It measures dispersion in the same unit as the data, making it more interpretable than variance.

Example : Standard Deviation = sqrt(8) = 2.83



**4. What is a box plot, and what can it tell you about the ddistribution of data?**


---
A box plot is a graphical representation of the distribution of a dataset that shows its spread, central tendency, and variability. It is particularly useful for visualizing the five-number summary of the data:

* Minimum: The smallest value in the dataset.

* First Quartile (Q1): The 25th percentile; the median of the lower half of the data.

* Median (Q2): The 50th percentile; the middle value of the dataset.

* Third Quartile (Q3): The 75th percentile; the median of the upper half of the data.

* Maximum: The largest value in the dataset.

A Box Plot can tell the following about the distribution :    

* **Central Tendency:**

    * The median line inside the box shows where the center of the data lies.
* **Spread (Variability):**

    * The IQR (length of the box) indicates the spread of the middle 50% of the data.
    * A larger IQR means greater variability.
* **Skewness:**

    * If the median is closer to Q1, the data is positively skewed (longer tail on the right).
    * If the median is closer to Q3, the data is negatively skewed (longer tail on the left).
* **Outliers:**

    * Points outside the whiskers highlight potential outliers or unusual values in the dataset.
* Range:

    * The length of the whiskers shows the overall range of the data (excluding outliers).





**5. Discuss the role of random sampling in making inference about populations.**


---
**ROLE OF RANDOM SAMPLING**

1. **Estimates of Population Parameters:**
    * A random sample provides data to estimate population parameters such as the mean, proportion, variance, etc.
    * Example: A random sample of 1,000 voters is surveyed to estimate the proportion of people supporting a candidate.
2. **Confidence Intervals:**
    * Random sampling allows the calculation of confidence intervals, which provide a range of values within which the true population parameter likely lies.
    * Example: A 95% confidence interval for the average income in a city might be $50,000–$55,000, meaning there’s a 95% chance the true average lies within this range.
3. **Hypothesis Testing:**
   * Random sampling ensures the validity of statistical tests by satisfying assumptions like independence and randomness.
   * Example: Testing whether the mean weight of a population is different from a hypothesized value.
4. **Reduces Sampling Error:**
   * Random sampling minimizes the chance that the sample is unrepresentative of the population, reducing sampling error.
   * Example: Sampling only one neighborhood to estimate city-wide income could lead to biased results, but random sampling from all neighborhoods provides a better estimate.




**6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?**


---
**SKEWNESS** is a measure of the asymmetry of a dataset's distribution. In a perfectly symmetrical dataset (e.g., a normal distribution), the left and right sides of the distribution are mirror images. When the distribution is not symmetrical, it is said to be skewed.

**TYPES OF SKEWNESS**

**1. Positive Skewness (Right-Skewed):**

* The tail on the right side (higher values) of the distribution is longer than the left.
* The mean is typically greater than the median.
* Examples:
    * Income distribution: Most people earn average or below-average income, with fewer people earning extremely high incomes.
    * Real estate prices: Many affordable properties with fewer high-value properties.

**2. Negative Skewness (Left-Skewed):**

* The tail on the left side (lower values) of the distribution is longer than the right.
* The mean is typically less than the median.
* Examples:
    * Test scores: If most students perform well, but a few score very poorly.
    * Age at retirement: Most people retire at a typical age, with fewer retiring significantly earlier.

**3. Zero Skewness (Symmetrical):**

* The distribution is perfectly symmetrical, with the mean, median, and mode being equal.
* Examples:
    * Heights of adults within a specific population (approximately symmetrical).
    * Errors in measurements that follow a normal distribution.


**IMPACT OF SKEWNESS ON DATA**

**1. Central Tendency:**
* Symmetrical Data: Mean, median, and mode are approximately the same.

* Skewed Data:
    * Mean is pulled in the direction of the skew (toward the longer tail).
    * Median remains closer to the center of the data and is less affected by extreme values.
    * Mean>Median>Mode.

**2.Spread and outliers:**
* Skewed distributions often indicate the presence of outliers in the direction of the skew.
* Example: In a right-skewed income distribution, a few extremely high incomes may disproportionately influence the mean.



**7. What is the interquartile range(IQR), and how is it used to detect outliers?**


---
**INTERQUARTILE RANGE(IQR)** is a measure of statistical dispersion that describes the spread of the middle 50% of the dataset. It is calculated as the difference between the **third quartile**(Q3) and the **first quartile**(Q1).

<center><b>IQR = Q3 - Q1</b></center><br>

**USING IQR TO DETECT OUTLIERS**<br>
Outliers are values that fall significantly outside the overall pattern of the data. The IQR is used to identify outliers through the following rule:<br>
**Step 1: Define lower and upper bounds**
* Lower Bound : Q1 - 1.5 * IQR
* Upper Bound : Q3 + 1.5 * IQR

**Step 2: Identify Outliers**
* Any data point below the lower bound or above the upper bound is considered an outlier.





**8. Discuss the conditions under which the binomial distribution is used.**


---
**BINOMIAL DISTRIBUTION** is a probability distribution that represents number of favourable outcomes in a fixed number of independent trials of an experiment, where only two outcomes of the trial are possible : true or false/ success or failure.
<br><br>
**CONDITIONS FOR USING THE BINOMIAL DISTRIBUTION**

**1. Fixed Numbers of Trials(n):**
* The experiment must consist of a predetermined, fixed number of trials.
* eg : Tossing a coin 10 times.(n=10)

**2. Two Outcomes per Trial:**
* Each trial must have exactly two outcomes, commonly labeled as success and failure or true and false.
* eg: Flipping a coin has 2 outcomes only(mutually exclusive) i.e heads or tails.

**3. Constant Probability of Success:**

* The probability of success must remain constant for all trials.
* eg: The probability of rolling a 6 on a fair die (p=1/6) does not change between rolls.

**4. Independence of Trials:**

* The outcome of one trial must not affect the outcomes of other trials.
* eg: In quality control, testing whether a product is defective should not influence the next product's result.

**5. Discrete Random Variable:**
* The random variable of interest (X) must count the number of successes (X=0,1,2,…,n).




**9. Explain the properties of the normal distribution and empirical rule(68-95-99.7 rule).**


---
**PROPERTIES OF NORMAL DISTRIBUTION**

**1. Shape:**
* The normal distribution is symmetric about its mean(x).
* It has a bell-shaped curve, where the highest point is at the mean.

**2. Mean, Median, mode:**
* For a normal distribution, mean, median and mode are all equal and located at the centre of distribution.

**3. 2 Parameters:**
* Mean(x) : Determines centre of the distribution.
* Standard Deviation(a) : Determines the spread of distribution.

**4. Tails Extend Infinitely**

**5.  Total Area under the curve is 1(total probability).**

**6. Curve is perfectly symmetrical about mean(x).**
<br><br>
**Empirical Rule(68-95-99.7)**

1. Approx. 68% of data falls within 1 standard deviation.(x +/- a)
2. Approx. 95% of data falls within 2 standard deviation.(x +/- 2a)
3. Approx. 99.7% of data falls within 3 standard deviation.(x +/- 3a)

**11. Explain what a random variable is and differentiate between discrete and continuous random variables.**


---
**RANDOM VARIABLE** is a variable that takes on numerical values based on the outcomes of a random experiment. It assigns a number to each possible outcome of a random process or event.

* Random variables are a key concept in probability and statistics, used to model uncertainty and variability.
* They are typically denoted by uppercase letters such as X, Y, or Z.
* The specific outcomes that the variable can take are called realizations or values, often denoted by lowercase letters like x, y or z.
<br>

**TYPES of Random Variables and their differentiation:**
1. Discrete Random Variable
    * Takes on countable values.(whole numbers)
    * Nature of values is Finite or countably infinite.
    * It uses probability mass function(PMF).
    * Probability of exact value is greater than zero for a specific value.
    * eg : No. of students in a class(n = 30).
2. Continuous Random Variable
    * Takes on uncountable continuous values(real numbers).
    * Nature of values is infinite and continuous within an interval.
    * It uses probability density function(PDF).
    * Probability of exact value is always zero for an excat value calculated over an interval.
    * eg : Height of students in a class(m = 170.5 cm)
